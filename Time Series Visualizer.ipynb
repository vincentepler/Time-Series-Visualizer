```python
## 📊 Project Overview: Page View Time Series Visualizer

This project is part of the freeCodeCamp Data Analysis with Python curriculum. The goal is to **visualize time series data** using Matplotlib, Pandas, and Seaborn.

This project used dataset containing the **daily page views** of the [freeCodeCamp.org] forum from **May 2016 to December 2019**. The data was cleaned and visualized in three different formats to uncover trends and patterns.

### 🔍 Objectives:
- Clean the data by removing outliers.
- Create the following plots:
  - **Line Plot**: Show the daily evolution of page views.
  - **Bar Plot**: Display monthly averages for each year.
  - **Box Plots**:
    - **Year-wise** box plot to reveal trends.
    - **Month-wise** box plot to reveal seasonality.

### 🧰 Tools Used:
- **Python**
- **Pandas** (for data cleaning and transformation)
- **Matplotlib** and **Seaborn** (for visualization)

### 📁 Dataset:
- File: `fcc-forum-pageviews.csv`
- Columns:
  - `date`: Date (index)
  - `value`: Number of page views
```

```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters

# Enable date converters for matplotlib
register_matplotlib_converters()

# Load dataset
df = pd.read_csv("fcc-forum-pageviews.csv", parse_dates=["date"], index_col="date")

# Clean data by removing the top and bottom 2.5% of page views
df = df.loc[
    (df['value'] >= df['value'].quantile(0.025)) &
    (df['value'] <= df['value'].quantile(0.975))
]
```

```python
def draw_line_plot():
    fig, ax = plt.subplots(figsize=(15, 5))
    ax.plot(df.index, df['value'], color='red', linewidth=1)

    # Title and labels
    ax.set_title("Daily freeCodeCamp Forum Page Views 5/2016 - 12/2019")
    ax.set_xlabel("Date")
    ax.set_ylabel("Page Views")

    # Save and return figure
    fig.savefig("line_plot.png")
    return fig

# Display the plot
draw_line_plot()
```

```python
def draw_bar_plot():
    # Prepare monthly average data
    df_bar = df.copy()
    df_bar["year"] = df_bar.index.year
    df_bar["month"] = df_bar.index.month

    df_grouped = df_bar.groupby(["year", "month"])["value"].mean().unstack()

    # Plot
    fig = df_grouped.plot(
        kind="bar",
        figsize=(12, 6),
        ylabel="Average Page Views",
        xlabel="Years"
    ).figure

    plt.legend(
        title="Month",
        labels=[
            "January", "February", "March", "April", "May", "June",
            "July", "August", "September", "October", "November", "December"
        ]
    )
    plt.xticks(fontsize=8)
    plt.yticks(fontsize=8)

    # Save and return figure
    fig.savefig("bar_plot.png")
    return fig

# Display the plot
draw_bar_plot()
```

```python
def draw_box_plot():
    # Prepare data
    df_box = df.copy()
    df_box.reset_index(inplace=True)
    df_box["year"] = df_box["date"].dt.year
    df_box["month"] = df_box["date"].dt.strftime('%b')
    df_box["month_num"] = df_box["date"].dt.month
    df_box = df_box.sort_values("month_num")

    # Create subplots
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))

    # Year-wise box plot with distinct colors
    sns.boxplot(
        x="year",
        y="value",
        data=df_box,
        ax=axes[0],
        hue="year",
        palette="Set2",
        dodge=False,
        legend=False
    )
    axes[0].set_title("Year-wise Box Plot (Trend)")
    axes[0].set_xlabel("Year")
    axes[0].set_ylabel("Page Views")

    # Month-wise box plot with distinct colors
    sns.boxplot(
        x="month",
        y="value",
        data=df_box,
        ax=axes[1],
        hue="month",
        palette="tab10",
        dodge=False,
        legend=False
    )
    axes[1].set_title("Month-wise Box Plot (Seasonality)")
    axes[1].set_xlabel("Month")
    axes[1].set_ylabel("Page Views")

    # Save and return figure
    fig.savefig("box_plot.png")
    return fig

# Display the updated plot
draw_box_plot()
```
